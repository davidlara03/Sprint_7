```python
import pandas as pd 
import plotly.express as px 

df= pd.read_csv("C:/Users/larag/Sprint_7/vehicles_us.csv")


print("Primeras filas del dataset:")
display(df.head())

print("\nInformación general:")
df.info()
```

```python
print("\nValores nulos por columna:")
print(df.isnull().sum())
```

```python
# Convertir date_posted a tipo datetime
df['date_posted'] = pd.to_datetime(df['date_posted'])

# Convertir is_4wd a tipo booleano: True si es 1.0, False si es NaN
df['is_4wd'] = df['is_4wd'].fillna(0).astype(bool)

# Eliminar filas con valores faltantes críticos: model_year, cylinders, odometer
df.dropna(subset=['model_year', 'cylinders', 'odometer'], inplace=True)

# Rellenar paint_color faltante con 'unknown'
df['paint_color'] = df['paint_color'].fillna('unknown')

# Convertir model_year a entero después de eliminar nulos
df['model_year'] = df['model_year'].astype(int)

# Convertir cylinders a entero
df['cylinders'] = df['cylinders'].astype(int)

print(df.isna().sum())
```

```python
fig = px.histogram(df, x='price', nbins=50, title='Distribución de Precios de Vehículos')

fig.show()
```

```python
fig = px.scatter(df, x='odometer', y='price',
                 title='Relación entre Precio y Kilometraje',
                 labels={'odometer': 'Kilometraje', 'price': 'Precio'},
                 opacity=0.6)
fig.show()
```
